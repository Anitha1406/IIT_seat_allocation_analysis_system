```python
def find_new_branches_with_colleges(df):
    # Sort the data by year for proper comparison
    df = df.sort_values(by='Year')
    
    # Dictionary to hold new branches and colleges added each year
    new_branches_with_colleges = {}
    
    # Get unique (Institute, Academic Program Name) pairs for the first year
    previous_year_data = set(df[df['Year'] == df['Year'].min()][['Institute', 'Academic_Program_Name']].apply(tuple, axis=1))
    
    # Loop through each year starting from the second year
    for year in sorted(df['Year'].unique())[1:]:
        # Get unique (Institute, Academic Program Name) pairs for the current year
        current_year_data = set(df[df['Year'] == year][['Institute', 'Academic_Program_Name']].apply(tuple, axis=1))
        
        # Find the difference (new branches with colleges added this year)
        new_in_year = current_year_data - previous_year_data
        
        if new_in_year:
            new_branches_with_colleges[year] = new_in_year
        
        # Update previous_year_data for the next iteration
        previous_year_data = previous_year_data.union(current_year_data)
    
    return new_branches_with_colleges

# Find new branches along with colleges added through the years
new_branches_colleges = find_new_branches_with_colleges(df)

# Display the result
for year, branches in new_branches_colleges.items():
    print(f"New branches added in {year}: (Total = {len(new_branches_colleges[year])})\n")
    for institute, branch in branches:
        print(f"{branch} at {institute}")
    print('\n')
```
